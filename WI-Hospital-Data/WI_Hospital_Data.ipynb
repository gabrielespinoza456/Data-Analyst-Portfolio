# Cleaning Healthcare Data

```python
import pandas as pd

# In order to properly visualize the hospitals on a map, we need their latitudes and longitudes
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
pd.set_option('display.max_columns', None)
pd.set_option('display.precision', 3)
```

```python
df = pd.read_csv('Hospital_General_Information.csv')
df
```

```python
# Dropping duplicate rows (just in case any exist) and filtering to only show hospitals in Wisconsin
df.drop_duplicates(inplace=True)
df2 = df[df['State'] == 'WI']
df2
```

```python
# Filtering out rows where scores are not available
# Although we only filtered on three columns, this ended up clearing all rows with 'Not Available' scores

df3 = df2[
        (df2['Hospital overall rating'] != 'Not Available')
        & (df2['Count of READM Measures Better'] != 'Not Available')
        & (df2['Count of READM Measures Worse'] != 'Not Available')
    ]
df3
```

```python
# Some of the numeric columns are not considered numeric in this dataset. Let's fix that
df3.dtypes
```

```python
cols_to_numeric = [
    'Hospital overall rating',
    'Count of Facility MORT Measures',
    'Count of MORT Measures Better',
    'Count of MORT Measures No Different',
    'Count of MORT Measures Worse',
    'Count of Facility Safety Measures',
    'Count of Safety Measures Better',
    'Count of Safety Measures No Different',
    'Count of Safety Measures Worse',
    'Count of Facility READM Measures',
    'Count of READM Measures Better',
    'Count of READM Measures No Different',
    'Count of READM Measures Worse'
]

df4 = df3.copy()
df4[cols_to_numeric] = df3[cols_to_numeric].apply(pd.to_numeric, errors='coerce')
df4
```

```python
# Dropping all the footnote columns since we won't need them for analysis

df5 = df4.copy()
df5 = df5.loc[:, ~df.columns.str.contains('footnote', case=False)]
df5
```

```python
df5.groupby('Hospital Ownership').mean(numeric_only=True)
```

```python
# Creating new columns for our Tableau dashboard
# These columns will have net scores which is (measures better) - (measures worse)
df6 = df5.copy()

# The Full Address column  will be used to find the latitude and longitude
df6['Full Address'] = df6['Address'] + ', ' + df6['City/Town'] + ', ' + df6['State']
df6['Net MORT Score'] = df6['Count of MORT Measures Better'] - df6['Count of MORT Measures Worse']
df6['Net Safety Score'] = df6['Count of Safety Measures Better'] - df6['Count of Safety Measures Worse']
df6['Net READM Score'] = df6['Count of READM Measures Better'] - df6['Count of READM Measures Worse']
```

```python
# Making sure the columns were added properly
df6
```

```python
df6.groupby('Hospital Ownership').mean(numeric_only=True)
```

```python
# All of the non 'Y' values in this column are NaN, meaning we can safely replace these with N
df6['Meets criteria for birthing friendly designation'].values
```

```python
df6['Meets criteria for birthing friendly designation'].fillna('N', inplace=True)
df6['Meets criteria for birthing friendly designation'].values
```

```python
df6
```

```python
# Retrieving the longitude / latitude of each location

geolocator = Nominatim(user_agent='WI_Hospital_Data_Coordinates', timeout=10)
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=5, max_retries=5) # Adding a delay as to not be flagged by the API
```

```python
# Creating a column of Location objects to extract the latitude/longitude from
# This may take a few minutes to run
df6['Location'] = df6['Full Address'].apply(geocode)
```

```python
# Adding the latitude and longitude
df6['Latitude'] = df6['Location'].apply(lambda x: x.latitude if x else None)
df6['Longitude'] = df6['Location'].apply(lambda x: x.longitude if x else None)
```

```python
# We have four rows with NaN coordinates. We can fill these in manually
df6[(df6['Latitude'].isna()) | (df6['Longitude'].isna())]
```
